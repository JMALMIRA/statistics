Para dat3, si asumimos que son muestras de X, una Bin(n,p), sabemos que 
        $$E(X)=np, \quad         V(X)=np(1-p)$$ 
Por tanto 
$$1-p = V(X)/E(X)$$ 
$$p = 1-V(X)/E(X)$$

```python
p=1-dat3.var()/dat3.mean()
p
```

Una vez conocida p, 
$$n = E(X)/p$$

```python
n=dat3.mean()/p
n
```

Cogemos $n = 146$ y $p=0.5$ Vamos a ver hasta qué punto se parecen los datos a la v.a. Binomial con los parámetros que acabamos de estimar:

```python
from scipy.stats import binom
X3=binom(146,p)
```

```python
# Q-Q plot para dat1 y dat2
Q = np.linspace(0.1, 0.9, 9)
Qdat3 = np.percentile(dat3, Q*100)
Qdat3teor = X3.ppf(Q)

plt.figure(figsize=(6, 6))
plt.plot(Qdat3, Qdat3teor, 'bo', label="Percentiles dat3 vs X3")
plt.plot([min(Qdat3), max(Qdat3)], [min(Qdat3), max(Qdat3)], 'r--', label='y = x')
plt.title('Q-Q plot: dat3 vs X3')
plt.xlabel('Quantiles de dat3')
plt.ylabel('Quantiles de X3')
plt.legend()
plt.grid(True)
plt.show()
```

Son muy parecidos. Nuestros datos dat3 se parecen mucho a la distribución binomial con n=146 y p = 0.51.


Para dat4, sabemos que, al ser la variable geométrica de parámetro $p$, $E(X) = (1-p)/p$, luego $p= 1/(1+media)$ (que sale de despejar). 

```python
p=1/(1+dat4.mean())
p
```


Otro modo podría ser tener en cuenta la esperanza y la varianza a la vez, pues: $E(X)=(1-p)/p$, $V(X)=(1-p)/p^2=E(X)/p$, implican que $p=E(X)/V(X)$, de modo que $p=E(X)/V(X)$.


```python
p=dat4.mean()/dat4.var()
p
```
